## D) Le vent : Calculs et manipulation de séries temporelles

### Étape 1 : Import des bibliothèques

```python
# Bibliothèques déjà importées
```

### Étape 2 : Import de l'ensemble de données

```python
# Chargement du fichier vent.csv
vent_raw = pd.read_csv('C:/Users/hp/Documents/AS3_2025-2026/SEMESTRE_1/Programmation Python/Projet_Python_AS3/tp_python AS3/vent.csv', sep=';')

print("Données de vent chargées")
print(f"Dimensions : {vent_raw.shape}")
print(vent_raw.head())
```

### Étape 3 : Attribution à data avec index datetime

```python
# Copie des données
data = vent_raw.copy()

# Conversion en entiers
data['Yr'] = data['Yr'].astype(int)
data['Mo'] = data['Mo'].astype(int)
data['Dy'] = data['Dy'].astype(int)

# Ajout de 2000 pour avoir un format année à 4 chiffres 
data['Year_temp'] = 2000 + data['Yr']

# Création d'un index datetime à partir des colonnes Yr, Mo, Dy
data['Date'] = pd.to_datetime(data[['Year_temp', 'Mo', 'Dy']].rename(
    columns={'Year_temp': 'year', 'Mo': 'month', 'Dy': 'day'}))

# Définition de l'index et suppression des colonnes d'origine
data = data.set_index('Date')
data = data.drop(['Yr', 'Mo', 'Dy', 'Year_temp'], axis=1)

print("Index datetime créé :")
print(data.head())
print(f"\nType de l'index : {type(data.index)}")
```

### Étape 4 : Correction de l'année 2061

```python
print("\nAnnée 2061? Avons-nous vraiment des données de cette année?")
print("Non, ce sont des données de 1961-1978 mal interprétées!")

# Fonction de correction pour l'année
def corriger_annee(date):
    """
    Corrige les dates des années 20XX en 19XX.
    """
    if date.year > 2000:
        return date.replace(year=date.year - 100)
    else:
        return date

# Application de la fonction
data.index = data.index.map(corriger_annee)

print("\nAprès correction :")
print(sorted(data.index.year.unique()))
print("\nPremières lignes :")
print(data.head())
```

### Étape 5 : Vérification du type datetime64[ns]

```python
print(f"Type de l'index : {data.index.dtype}")
print(f"\nL'index est bien de type datetime64[ns] : {data.index.dtype == 'datetime64[ns]'}")
data.head()
```

### Étape 6 : Calcul des valeurs manquantes

```python
# Nombre de valeurs manquantes par variable
valeurs_manquantes = data.isnull().sum()

print("Nombre de valeurs manquantes par localisation :")
print(valeurs_manquantes)

print("\nTotal de valeurs manquantes :")
print(f"Total : {valeurs_manquantes.sum()}")

print("\nPourcentage de valeurs manquantes par localisation :")
pourcentage_manquantes = (valeurs_manquantes / len(data)) * 100
print(pourcentage_manquantes.sort_values(ascending=False))
```

### Étape 7 : Calcul du nombre total de valeurs non manquantes

```python
# Nombre de valeurs non manquantes
valeurs_non_manquantes = data.notnull().sum().sum()

print(f"Nombre total de valeurs non manquantes : {valeurs_non_manquantes}")

# Calculs complémentaires
total_cellules = data.shape[0] * data.shape[1]
print(f"\nNombre total de cellules : {total_cellules}")
print(f"Valeurs manquantes : {valeurs_manquantes.sum()}")
print(f"Valeurs non manquantes : {valeurs_non_manquantes}")
print(f"\nPourcentage de données disponibles : {(valeurs_non_manquantes / total_cellules) * 100:.2f}%")
```

### Étape 8 : Vitesse moyenne globale du vent

```python
# Calcul de la moyenne globale (tous emplacements, toutes heures)
vitesse_moyenne_globale = data.mean().mean()

print(f"Vitesse moyenne globale du vent : {vitesse_moyenne_globale:.2f}")

print("\nMoyennes par localisation :")
print(data.mean().sort_values(ascending=False))
print("\n Moyennes par heure :")
print(data.groupby(data.index.hour).mean().sort_values(by=data.columns[0], ascending=False))
```

### Étape 9 : DataFrame loc_stats (statistiques par localisation)

```python
# Calcul des statistiques par localisation
loc_stats = pd.DataFrame({
    'min': data.min(),
    'max': data.max(),
    'mean': data.mean(),
    'std': data.std()
})

print("Statistiques par localisation :")
print(loc_stats)

print("\nLocalisation la plus venteuse (moyenne la plus élevée) :")
print(loc_stats['mean'].idxmax(), f": {loc_stats['mean'].max():.2f} nœuds")

print("\nLocalisation la moins venteuse (moyenne la plus faible) :")
print(loc_stats['mean'].idxmin(), f": {loc_stats['mean'].min():.2f} nœuds")
```

### Étape 10 : DataFrame day_stats (statistiques par jour)

```python
# Calcul des statistiques par jour
day_stats = pd.DataFrame({
    'min': data.min(axis=1),
    'max': data.max(axis=1),
    'mean': data.mean(axis=1),
    'std': data.std(axis=1)
})

print("Statistiques par jour (10 premières lignes) :")
print(day_stats.head(10))

print("\nJour le plus venteux :")
jour_max = day_stats['mean'].idxmax()
print(f"{jour_max.strftime('%Y-%m-%d')} : {day_stats.loc[jour_max, 'mean']:.2f} nœuds")

print("\nJour le moins venteux :")
jour_min = day_stats['mean'].idxmin()
print(f"{jour_min.strftime('%Y-%m-%d')} : {day_stats.loc[jour_min, 'mean']:.2f} nœuds")
```

### Étape 11 : Vitesse moyenne du vent en janvier

```python
# Filtrage des données de janvier 1961 et janvier 1962 
janvier = data[(data.index.month == 1) & (data.index.year.isin([1961, 1962]))]

# Calcul de la moyenne par localisation pour janvier
moyenne_janvier = janvier.mean()

print("Vitesse moyenne du vent en janvier par localisation :")
print(moyenne_janvier.sort_values(ascending=False))

print(f"\nNombre de jours de janvier analysés : {len(janvier)}")
print(f"Années considérées : 1961 et 1962")
print(f"Vitesse moyenne globale en janvier : {moyenne_janvier.mean():.2f} nœuds")
```

### Étape 15 : Statistiques hebdomadaires (52 premières semaines)

```python
# Filtrage des 52 premières semaines à partir du 2 janvier 1961
date_debut = pd.Timestamp('1961-01-02')
date_fin = date_debut + pd.Timedelta(weeks=52)

data_52_semaines = data[date_debut:date_fin]

# Regroupement par semaine et calcul des statistiques
stats_hebdo = data_52_semaines.resample('W').agg(['min', 'max', 'mean', 'std'])

print("Statistiques hebdomadaires (premières semaines) :")
print(stats_hebdo.head(10))

print(f"\nNombre de semaines analysées : {len(stats_hebdo)}")

# Statistiques globales hebdomadaires
print("\nMoyenne des statistiques hebdomadaires sur toutes les localisations :")
moyennes_hebdo = pd.DataFrame({
    'min': stats_hebdo.xs('min', level=1, axis=1).mean(axis=1),
    'max': stats_hebdo.xs('max', level=1, axis=1).mean(axis=1),
    'mean': stats_hebdo.xs('mean', level=1, axis=1).mean(axis=1),
    'std': stats_hebdo.xs('std', level=1, axis=1).mean(axis=1)
})
print(moyennes_hebdo.head())
```

   - Ces outliers écrasent l'échelle du graphique

   - Utilisation de `mean(UnitPrice)` = moyenne SIMPLE
   - **Problème :** Ne reflète pas le prix réellement payé par article
   
   **Exemple :**
   - Client achète : 1 article à 10£ + 100 articles à 2£
   - Moyenne simple : (10 + 2) / 2 = **6£**
   - Prix moyen réel : (1×10 + 100×2) / 101 = **2.08£**

4. **ÉCHELLE INADAPTÉE**
   - Axe Y : 0 → 200,000
   - 95% des données invisibles entre 0-10,000

### Étape 8.1 : Groupement du prix unitaire par intervalles et agrégation

```python
# Création des intervalles de prix [0, 1), [1, 2), ..., [49, 50)
online_rt_positif['PriceRange'] = pd.cut(online_rt_positif['UnitPrice'], 
                                          bins=range(0, 51, 1), 
                                          right=False)

# Agrégation par intervalle de prix
price_analysis = online_rt_positif.groupby('PriceRange', observed=True).agg({
    'Quantity': 'sum',
    'Revenue': 'sum'
}).reset_index()

# Suppression des intervalles sans données
price_analysis = price_analysis[price_analysis['Quantity'] > 0]

print("Analyse par tranche de prix :")
print(price_analysis.head(15))

print("\nStatistiques :")
print(f"Nombre d'intervalles avec des ventes : {len(price_analysis)}")
print(f"Revenu total : £{price_analysis['Revenue'].sum():,.2f}")
print(f"Quantité totale : {price_analysis['Quantity'].sum():,}")
```

### Étape 8.3 : Tracé des graphiques

```python
# Extraction de la borne inférieure de chaque intervalle
price_analysis['PriceLower'] = price_analysis['PriceRange'].apply(lambda x: x.left)

# Création de deux sous-graphiques
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(16, 10))

# Graphique 1 : Quantité par intervalle de prix
ax1.plot(price_analysis['PriceLower'], price_analysis['Quantity'], 
         linewidth=2.5, color='#2E86AB', marker='o', markersize=6, 
         markerfacecolor='#A23B72', markeredgecolor='black', markeredgewidth=0.5)
ax1.fill_between(price_analysis['PriceLower'], price_analysis['Quantity'], 
                 alpha=0.3, color='#2E86AB')
ax1.set_xlabel('Prix unitaire (£)', fontsize=12, fontweight='bold')
ax1.set_ylabel('Quantité totale', fontsize=12, fontweight='bold')
ax1.set_title('Quantité vendue par tranche de prix', fontsize=14, fontweight='bold', pad=15)
ax1.grid(True, alpha=0.3, linestyle='--')

# Graphique 2 : Revenu par intervalle de prix
ax2.plot(price_analysis['PriceLower'], price_analysis['Revenue'], 
         linewidth=2.5, color='#F18F01', marker='s', markersize=6,
         markerfacecolor='#C73E1D', markeredgecolor='black', markeredgewidth=0.5)
ax2.fill_between(price_analysis['PriceLower'], price_analysis['Revenue'], 
                 alpha=0.3, color='#F18F01')
ax2.set_xlabel('Prix unitaire (£)', fontsize=12, fontweight='bold')
ax2.set_ylabel('Revenu total (£)', fontsize=12, fontweight='bold')
ax2.set_title('Revenu généré par tranche de prix', fontsize=14, fontweight='bold', pad=15)
ax2.grid(True, alpha=0.3, linestyle='--')

plt.tight_layout()
plt.show()

# ============================================================================
# GRAPHIQUE SUPPLÉMENTAIRE : Superposition avec deux axes Y
# ============================================================================

fig, ax1 = plt.subplots(figsize=(16, 8))

# Axe Y gauche : Quantité
color_quantity = '#2E86AB'
ax1.set_xlabel('Prix unitaire (£)', fontsize=13, fontweight='bold')
ax1.set_ylabel('Quantité totale', fontsize=13, fontweight='bold', color=color_quantity)
line1 = ax1.plot(price_analysis['PriceLower'], price_analysis['Quantity'], 
                 linewidth=3, color=color_quantity, marker='o', markersize=7, 
                 markerfacecolor='#A23B72', markeredgecolor='white', markeredgewidth=1.5,
                 label='Quantité', zorder=3)
ax1.fill_between(price_analysis['PriceLower'], price_analysis['Quantity'], 
                 alpha=0.2, color=color_quantity, zorder=1)
ax1.tick_params(axis='y', labelcolor=color_quantity, labelsize=11)
ax1.grid(True, alpha=0.3, linestyle='--', zorder=0)

# Axe Y droit : Revenu
ax2 = ax1.twinx()
color_revenue = '#F18F01'
ax2.set_ylabel('Revenu total (£)', fontsize=13, fontweight='bold', color=color_revenue)
line2 = ax2.plot(price_analysis['PriceLower'], price_analysis['Revenue'], 
                 linewidth=3, color=color_revenue, marker='s', markersize=7,
                 markerfacecolor='#C73E1D', markeredgecolor='white', markeredgewidth=1.5,
                 label='Revenu', zorder=2)
ax2.tick_params(axis='y', labelcolor=color_revenue, labelsize=11)

# Titre et légende
plt.title('Quantité et Revenu par tranche de prix unitaire', 
          fontsize=15, fontweight='bold', pad=20)

# Légende combinée
lines = line1 + line2
labels = [l.get_label() for l in lines]
ax1.legend(lines, labels, loc='upper right', fontsize=12, framealpha=0.9)

plt.tight_layout()
plt.show()

print("\n Analyse du graphique superposé :")
print("- La courbe BLEUE (Quantité) montre le volume vendu par tranche de prix")
print("- La courbe ORANGE (Revenu) montre le chiffre d'affaires généré")
print("- Les deux courbes permettent de voir la relation entre volume et revenu")
```

```python
# Configuration du style général
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

# Création de graphiques avec design premium
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(20, 14))

# Définition des palettes de couleurs
colors_qty = {
    'main': '#0984e3',
    'fill': '#74b9ff',
    'accent': '#d63031',
    'bg': '#ffeaa7'
}

colors_rev = {
    'main': '#00b894',
    'fill': '#55efc4',
    'accent': '#fdcb6e',
    'bg': '#dfe6e9'
}

# ============================================================================
# GRAPHIQUE 1 : Quantité par intervalle de prix
# ============================================================================

# Courbe principale avec ombre portée
ax1.plot(price_analysis['PriceLower'], price_analysis['Quantity'], 
         linewidth=4, color=colors_qty['main'], marker='o', markersize=10, 
         markerfacecolor=colors_qty['accent'], markeredgecolor='white', 
         markeredgewidth=2.5, label='Quantité totale', zorder=5,
         alpha=0.9, markevery=1)

# Remplissage avec gradient
ax1.fill_between(price_analysis['PriceLower'], price_analysis['Quantity'], 
                 alpha=0.3, color=colors_qty['fill'], zorder=1)

# Ligne de moyenne avec style amélioré
moyenne_qty = price_analysis['Quantity'].mean()
ax1.axhline(y=moyenne_qty, color=colors_qty['accent'], linestyle='--', 
            linewidth=2.5, alpha=0.7, 
            label=f'Moyenne: {int(moyenne_qty):,} unités', zorder=2)

# Zone de confiance (±20% de la moyenne)
ax1.fill_between(price_analysis['PriceLower'], 
                 moyenne_qty * 0.8, moyenne_qty * 1.2,
                 alpha=0.1, color=colors_qty['accent'], zorder=0)

# Axes et labels
ax1.set_xlabel('Prix unitaire (£)', fontsize=16, fontweight='bold', 
               labelpad=15, color='#2d3436')
ax1.set_ylabel('Quantité totale (unités)', fontsize=16, fontweight='bold', 
               labelpad=15, color='#2d3436')
ax1.set_title(' Quantité vendue par tranche de prix unitaire', 
              fontsize=18, fontweight='bold', pad=25, color='#2d3436')

# Grille améliorée
ax1.grid(True, alpha=0.25, linestyle='--', linewidth=1.3, color='#636e72')
ax1.set_xlim(-1, 51)
ax1.set_ylim(bottom=0)

# Annotation du maximum avec design premium
max_qty_idx = price_analysis['Quantity'].idxmax()
max_qty_price = price_analysis.loc[max_qty_idx, 'PriceLower']
max_qty = price_analysis.loc[max_qty_idx, 'Quantity']

ax1.annotate(
    f' MAXIMUM\n{int(max_qty):,} unités\n[{max_qty_price:.0f}£ - {max_qty_price+1:.0f}£]\n{max_qty/price_analysis["Quantity"].sum()*100:.1f}% du total',
    xy=(max_qty_price, max_qty), 
    xytext=(max_qty_price + 15, max_qty * 0.75),
    arrowprops=dict(
        arrowstyle='->', 
        color=colors_qty['accent'], 
        lw=3.5, 
        connectionstyle='arc3,rad=0.4',
        alpha=0.9
    ),
    fontsize=12, fontweight='bold', color='#2d3436',
    bbox=dict(
        boxstyle='round,pad=1', 
        facecolor=colors_qty['bg'], 
        edgecolor=colors_qty['accent'], 
        linewidth=3, 
        alpha=0.95
    ),
    ha='left', va='center'
)

# Marqueurs pour le TOP 3
top_3_qty = price_analysis.nlargest(3, 'Quantity')
for i, (idx, row) in enumerate(top_3_qty.iterrows(), 1):
    if i == 1:
        continue
    ax1.plot(row['PriceLower'], row['Quantity'], 
            marker='*', markersize=20, color=colors_qty['accent'], 
            markeredgecolor='white', markeredgewidth=2, zorder=6)
    ax1.text(row['PriceLower'], row['Quantity'] * 1.08, f"TOP {i}", 
            fontsize=10, fontweight='bold', ha='center', 
            color=colors_qty['accent'],
            bbox=dict(boxstyle='round,pad=0.3', facecolor='white', 
                     edgecolor=colors_qty['accent'], alpha=0.8))

# Légende personnalisée
legend1 = ax1.legend(loc='upper right', fontsize=13, framealpha=0.97,
                    edgecolor='#2d3436', fancybox=True, shadow=True)
legend1.get_frame().set_linewidth(2)

# ============================================================================
# GRAPHIQUE 2 : Revenu par intervalle de prix
# ============================================================================

revenue_millions = price_analysis['Revenue'] / 1_000_000

# Courbe principale
ax2.plot(price_analysis['PriceLower'], revenue_millions, 
         linewidth=4, color=colors_rev['main'], marker='D', markersize=10,
         markerfacecolor=colors_rev['accent'], markeredgecolor='white', 
         markeredgewidth=2.5, label='Revenu total', zorder=5,
         alpha=0.9, markevery=1)

# Remplissage
ax2.fill_between(price_analysis['PriceLower'], revenue_millions, 
                 alpha=0.3, color=colors_rev['fill'], zorder=1)

# Ligne de moyenne
moyenne_rev = revenue_millions.mean()
ax2.axhline(y=moyenne_rev, color=colors_rev['accent'], linestyle='--', 
            linewidth=2.5, alpha=0.7, 
            label=f'Moyenne: £{moyenne_rev:.2f}M', zorder=2)

# Zone de confiance
ax2.fill_between(price_analysis['PriceLower'], 
                 moyenne_rev * 0.8, moyenne_rev * 1.2,
                 alpha=0.1, color=colors_rev['accent'], zorder=0)

# Axes et labels
ax2.set_xlabel('Prix unitaire (£)', fontsize=16, fontweight='bold', 
               labelpad=15, color='#2d3436')
ax2.set_ylabel('Revenu total (Millions £)', fontsize=16, fontweight='bold', 
               labelpad=15, color='#2d3436')
ax2.set_title(' Revenu généré par tranche de prix unitaire', 
              fontsize=18, fontweight='bold', pad=25, color='#2d3436')

# Grille
ax2.grid(True, alpha=0.25, linestyle='--', linewidth=1.3, color='#636e72')
ax2.set_xlim(-1, 51)
ax2.set_ylim(bottom=0)

# Annotation du maximum
max_rev_idx = price_analysis['Revenue'].idxmax()
max_rev_price = price_analysis.loc[max_rev_idx, 'PriceLower']
max_rev = price_analysis.loc[max_rev_idx, 'Revenue'] / 1_000_000

ax2.annotate(
    f' MAXIMUM\n£{max_rev:.2f}M\n[{max_rev_price:.0f}£ - {max_rev_price+1:.0f}£]\n{price_analysis.loc[max_rev_idx, "Revenue"]/price_analysis["Revenue"].sum()*100:.1f}% du total',
    xy=(max_rev_price, max_rev), 
    xytext=(max_rev_price + 15, max_rev * 0.75),
    arrowprops=dict(
        arrowstyle='->', 
        color=colors_rev['accent'], 
        lw=3.5, 
        connectionstyle='arc3,rad=0.4',
        alpha=0.9
    ),
    fontsize=12, fontweight='bold', color='#2d3436',
    bbox=dict(
        boxstyle='round,pad=1', 
        facecolor=colors_rev['bg'], 
        edgecolor=colors_rev['accent'], 
        linewidth=3, 
        alpha=0.95
    ),
    ha='left', va='center'
)

# TOP 3 Revenu
top_3_rev = price_analysis.nlargest(3, 'Revenue')
for i, (idx, row) in enumerate(top_3_rev.iterrows(), 1):
    if i == 1:
        continue
    rev_val = row['Revenue'] / 1_000_000
    ax2.plot(row['PriceLower'], rev_val, 
            marker='*', markersize=20, color=colors_rev['accent'],
            markeredgecolor='white', markeredgewidth=2, zorder=6)
    ax2.text(row['PriceLower'], rev_val * 1.08, f"TOP {i}", 
            fontsize=10, fontweight='bold', ha='center', 
            color=colors_rev['accent'],
            bbox=dict(boxstyle='round,pad=0.3', facecolor='white', 
                     edgecolor=colors_rev['accent'], alpha=0.8))

# Légende
legend2 = ax2.legend(loc='upper right', fontsize=13, framealpha=0.97,
                    edgecolor='#2d3436', fancybox=True, shadow=True)
legend2.get_frame().set_linewidth(2)

# Ajustement final
plt.tight_layout(pad=3.0)
plt.show()

# ============================================================================
# STATISTIQUES DÉTAILLÉES AVEC FORMAT AMÉLIORÉ
# ============================================================================
print("\n" + "╔" + "="*78 + "╗")
print("║" + " "*20 + " ANALYSE DÉTAILLÉE DES TRANCHES DE PRIX" + " "*17 + "  ║")
print("╚" + "="*78 + "╝")

# Top 3 par quantité
print("\n TOP 3 - VOLUME DE VENTES (QUANTITÉ)")
print("┌" + "─"*78 + "┐")
for i, (idx, row) in enumerate(top_3_qty.iterrows(), 1):
    price_range = f"{row['PriceLower']:.0f}£ - {row['PriceLower']+1:.0f}£"
    pct = row['Quantity']/price_analysis['Quantity'].sum()*100
    bar = "█" * int(pct / 2)
    print(f"│ {i}. {price_range:12} → {int(row['Quantity']):>10,} unités │ {pct:>5.2f}% │ {bar:<30}│")
print("└" + "─"*78 + "┘")

# Top 3 par revenu
print("\n TOP 3 - CHIFFRE D'AFFAIRES (REVENU)")
print("┌" + "─"*78 + "┐")
for i, (idx, row) in enumerate(top_3_rev.iterrows(), 1):
    price_range = f"{row['PriceLower']:.0f}£ - {row['PriceLower']+1:.0f}£"
    rev_millions = row['Revenue'] / 1_000_000
    pct = row['Revenue']/price_analysis['Revenue'].sum()*100
    bar = "█" * int(pct / 2)
    print(f"│ {i}. {price_range:12} → £{rev_millions:>7.2f}M     │ {pct:>5.2f}% │ {bar:<30}    │")
print("└" + "─"*78 + "┘")

# Statistiques globales
print("\n INDICATEURS CLÉS")
print("┌" + "─"*78 + "┐")
total_rev = price_analysis['Revenue'].sum()/1_000_000
total_qty = price_analysis['Quantity'].sum()
avg_price = price_analysis['Revenue'].sum()/price_analysis['Quantity'].sum()
print(f"│ Revenu total              : £{total_rev:>8.2f} millions                               │")
print(f"│ Quantité totale           : {total_qty:>10,} unités                                │")
print(f"│ Prix moyen pondéré        : £{avg_price:>8.2f}                                        │")
print(f"│ Nombre de tranches actives: {len(price_analysis):>10}                                       │")
print("└" + "─"*78 + "┘")

# Analyse de concentration
rev_top_3 = top_3_rev['Revenue'].sum()
qty_top_3 = top_3_qty['Quantity'].sum()
print("\n ANALYSE DE CONCENTRATION (TOP 3)")
print("┌" + "─"*78 + "┐")
print(f"│ Concentration du revenu   : {rev_top_3/price_analysis['Revenue'].sum()*100:>5.1f}% du total                                 │")
print(f"│ Concentration du volume   : {qty_top_3/price_analysis['Quantity'].sum()*100:>5.1f}% du total                                 │")
print("└" + "─"*78 + "┘")

# Conclusion
print("\n" + "╔" + "="*78 + "╗")
print("║" + " "*28 + " INSIGHTS CLÉS" + " "*33 + "║")
print("╠" + "="*78 + "╣")
print(f"║ Tranche la plus rentable: {max_rev_price:.0f}£-{max_rev_price+1:.0f}£ (£{max_rev:.2f}M de revenu)" + " "*(78-len(f" Tranche la plus rentable: {max_rev_price:.0f}£-{max_rev_price+1:.0f}£ (£{max_rev:.2f}M de revenu)")-1) + "║")
print(f"║ Tranche avec plus de volume: {max_qty_price:.0f}£-{max_qty_price+1:.0f}£ ({int(max_qty):,} unités)" + " "*(78-len(f" Tranche avec plus de volume: {max_qty_price:.0f}£-{max_qty_price+1:.0f}£ ({int(max_qty):,} unités)")-1) + " ║")
print(f"║ Relation volume/revenu: {'Alignée' if max_qty_price == max_rev_price else 'Différente'}" + " "*(78-len(f"  Relation volume/revenu: {'Alignée' if max_qty_price == max_rev_price else 'Différente'}")-1) + "  ║")
print("╚" + "="*78 + "╝\n")
```

### Question bonus : Quel est le profil temporel des ventes et existe-t-il des tendances saisonnières ?

```python
# Analyse temporelle des ventes
online_rt_positif['Month'] = pd.to_datetime(online_rt_positif['InvoiceDate']).dt.month
online_rt_positif['DayOfWeek'] = pd.to_datetime(online_rt_positif['InvoiceDate']).dt.dayofweek
online_rt_positif['Hour'] = pd.to_datetime(online_rt_positif['InvoiceDate']).dt.hour

# Ventes par mois
ventes_mensuelles = online_rt_positif.groupby('Month')['Revenue'].sum()

# Ventes par jour de la semaine
jours = ['Lundi', 'Mardi', 'Mercredi', 'Jeudi', 'Vendredi', 'Samedi', 'Dimanche']
ventes_jour = online_rt_positif.groupby('DayOfWeek')['Revenue'].sum()
ventes_jour = ventes_jour.reindex(range(7), fill_value=0)

# Ventes par heure
ventes_heure = online_rt_positif.groupby('Hour')['Revenue'].sum()

# Création de trois graphiques
fig, axes = plt.subplots(1, 3, figsize=(20, 5))

# Graphique 1 : Ventes par mois
mois_noms = ['Jan', 'Fév', 'Mar', 'Avr', 'Mai', 'Jun', 
             'Jul', 'Aoû', 'Sep', 'Oct', 'Nov', 'Déc']

ventes_mensuelles = ventes_mensuelles.reindex(range(1, 13), fill_value=0)

axes[0].bar(range(1, 13), ventes_mensuelles.values, 
           color=sns.color_palette("coolwarm", 12), edgecolor='black', linewidth=1.2)
axes[0].set_xticks(range(1, 13))
axes[0].set_xticklabels(mois_noms, rotation=45)
axes[0].set_xlabel('Mois', fontsize=12, fontweight='bold')
axes[0].set_ylabel('Revenu total (£)', fontsize=12, fontweight='bold')
axes[0].set_title('Revenus par mois', fontsize=13, fontweight='bold')
axes[0].grid(axis='y', alpha=0.3, linestyle='--')

# Graphique 2 : Ventes par jour de la semaine
axes[1].bar(range(7), ventes_jour.values, 
           color=sns.color_palette("viridis", 7), edgecolor='black', linewidth=1.2)
axes[1].set_xticks(range(7))
axes[1].set_xticklabels(jours, rotation=45)
axes[1].set_xlabel('Jour de la semaine', fontsize=12, fontweight='bold')
axes[1].set_ylabel('Revenu total (£)', fontsize=12, fontweight='bold')
axes[1].set_title('Revenus par jour de la semaine', fontsize=13, fontweight='bold')
axes[1].grid(axis='y', alpha=0.3, linestyle='--')

# Graphique 3 : Ventes par heure
axes[2].plot(ventes_heure.index, ventes_heure.values, 
            linewidth=3, color='#E63946', marker='o', markersize=8,
            markerfacecolor='#F1FAEE', markeredgecolor='black', markeredgewidth=1)
axes[2].fill_between(ventes_heure.index, ventes_heure.values, alpha=0.3, color='#E63946')
axes[2].set_xlabel('Heure de la journée', fontsize=12, fontweight='bold')
axes[2].set_ylabel('Revenu total (£)', fontsize=12, fontweight='bold')
axes[2].set_title('Revenus par heure', fontsize=13, fontweight='bold')
axes[2].grid(True, alpha=0.3, linestyle='--')
axes[2].set_xticks(range(0, 24, 2))

plt.suptitle('Analyse temporelle des ventes en ligne', fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

# Affichage des résultats
print("\nRésultats de l'analyse temporelle :")
print("="*60)


print(f"\nMois le plus rentable : {mois_noms[ventes_mensuelles.idxmax() - 1]}")
if ventes_jour.max() > 0:
    print(f"Jour le plus rentable : {jours[ventes_jour.idxmax()]}")
print(f"Heure la plus rentable : {ventes_heure.idxmax()}h")
print(f"\nRevenu mensuel moyen : £{ventes_mensuelles.mean():,.2f}")
print(f"Revenu journalier moyen : £{ventes_jour[ventes_jour > 0].mean():,.2f}")
```
